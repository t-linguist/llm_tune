# 1. Загрузка датасета

```python
import pandas as pd
import pickle
```

```python
with open("data/test.pkl", "rb") as file2:
    test_data = pickle.load(file2)
test_data
```

```python
import numpy as np

SIZE = len(test_data)

en2es_idx = np.arange(0, SIZE, 6)
en2ru_idx = np.arange(1, SIZE, 6)
es2en_idx = np.arange(2, SIZE, 6)
es2ru_idx = np.arange(3, SIZE, 6)
ru2en_idx = np.arange(4, SIZE, 6)
ru2es_idx = np.arange(5, SIZE, 6)
```

```python
en2es_pairs = [(en2es_idx[0], en2es_idx[1])]
for idx in range(2, len(en2es_idx)):
    en2es_pairs.append((en2es_idx[idx-1], en2es_idx[idx]))
en2es_pairs += [(en2es_idx[-1], en2es_idx[-2])]

en2ru_pairs = [(en2ru_idx[0], en2ru_idx[1])]
for idx in range(2, len(en2ru_idx)):
    en2ru_pairs.append((en2ru_idx[idx-1], en2ru_idx[idx]))
en2ru_pairs += [(en2ru_idx[-1], en2ru_idx[-2])]

es2en_pairs = [(es2en_idx[0], es2en_idx[1])]
for idx in range(2, len(es2en_idx)):
    es2en_pairs.append((es2en_idx[idx-1], es2en_idx[idx]))
es2en_pairs += [(es2en_idx[-1], es2en_idx[-2])]

es2ru_pairs = [(es2ru_idx[0], es2ru_idx[1])]
for idx in range(2, len(es2ru_idx)):
    es2ru_pairs.append((es2ru_idx[idx-1], es2ru_idx[idx]))
es2ru_pairs += [(es2ru_idx[-1], es2ru_idx[-2])]

ru2en_pairs = [(ru2en_idx[0], ru2en_idx[1])]
for idx in range(2, len(ru2en_idx)):
    ru2en_pairs.append((ru2en_idx[idx-1], ru2en_idx[idx]))
ru2en_pairs += [(ru2en_idx[-1], ru2en_idx[-2])]

ru2es_pairs = [(ru2es_idx[0], ru2es_idx[1])]
for idx in range(2, len(ru2es_idx)):
    ru2es_pairs.append((ru2es_idx[idx-1], ru2es_idx[idx]))
ru2es_pairs += [(ru2es_idx[-1], ru2es_idx[-2])]

pairs_lst = en2es_pairs + en2ru_pairs + es2en_pairs + es2ru_pairs + ru2en_pairs + ru2es_pairs
pairs_lst
```

```python
data = [test_data[idx[0]]['instruction'] + test_data[idx[0]]['text'] + test_data[idx[0]]['translation'] + test_data[idx[1]]['text'] for idx in pairs_lst]
data
```

```python
len(data)
```

```python
y_true = [test_data[idx[1]]['translation'] for idx in pairs_lst]
y_true
```

```python
len(y_true)
```

# 2. Загрузка базовой модели

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

base_model_id = "mistralai/Mistral-7B-Instruct-v0.2"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

model = AutoModelForCausalLM.from_pretrained(base_model_id, quantization_config=bnb_config)
```

# 3. Токенизация

```python
tokenizer = AutoTokenizer.from_pretrained(
    base_model_id,
    padding_side="left",
    add_bos_token=True,
    add_eos_token=True,
    truncation=True,
    max_length=256,
    padding="max_length"
)

tokenizer.pad_token = tokenizer.eos_token
```

# 4. Инференс

```python
import re
from tqdm import tqdm

predictions = []
for sent in tqdm(data):
    model_input = tokenizer(sent, return_tensors="pt").to("cuda")
    model.eval()
    with torch.no_grad():
        result = tokenizer.decode(model.generate(**model_input, max_new_tokens=60, repetition_penalty=1.15)[0], skip_special_tokens=True)
        predictions.append(result)
predictions
```

```python
len(predictions)
```

```python
with open("mistral/one/preds.pkl", "wb") as file2:
    predictions = pickle.dump(file2)
predictions
```

```python
import re
predictions_clean = [re.findall("(###.*:\s{0,2})(.*)", predictions[n])[2][1] for n in range(len(predictions))]
predictions_clean
```

```python
for pred, tr in zip(predictions_clean, y_true):
    print(f"Предсказание: {pred}")
    print(f"Правильный ответ: {tr}")
    print("\n")
```

```python
with open("mistral/one/preds_postprocess.pkl", "wb") as file3:
    pickle.dump(predictions, file3)
with open("mistral/one/y_true.pkl", "wb") as file4:
    pickle.dump(y_true, file4)
```
